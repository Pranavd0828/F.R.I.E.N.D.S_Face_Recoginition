## Face Recognition:

```python
import PIL.Image
import PIL.ImageDraw
import face_recognition

from IPython.display import display, Image
```

Load the JPG file in numpy.array

```python
image_people = face_recognition.load_image_file("people.jpg")
```

```python
display(Image(filename='people.jpg'))
```

Call the face_location function from the face_recognition package. This will extract the location of the face.

```python
face_locations = face_recognition.face_locations(image_people)
number_of_faces = len(face_locations)
```

```python
print("No. of faces in the image = {}".format(number_of_faces))
```

```python
pil_image = PIL.Image.fromarray(image_people)
```

```python
for face_location in face_locations:
    
    top, right, bottom, left = face_location
    
    print("A face is located at pixel location- Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, 
                                                                                                bottom, right))

    draw = PIL.ImageDraw.Draw(pil_image)
    draw.rectangle([left, top, right, bottom], outline="blue", width = 5)
    
pil_image.show()
```

```python
face_landmarks_list = face_recognition.face_landmarks(image_people)
```

```python
for face_landmarks in face_landmarks_list:
    
    for name, list_of_face_features in face_landmarks.items():

        print("The {} in this face has following points: {}".format(name, list_of_face_features))
        draw.line(list_of_face_features, fill = "red", width = 5)
    
pil_image.show()
```
